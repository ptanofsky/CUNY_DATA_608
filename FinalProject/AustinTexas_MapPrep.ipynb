```python
import json

# 'dashapp/atx_zips_coords.json' ... contains just the 45 ATX zip codes
# data/tx_zips.json' ... contains all 1939 Texas zip codes
with open('dashapp/atx_zips_coords.json') as zips_file:
    zips_data = json.load(zips_file)
```

```python
# From: https://gis.stackexchange.com/questions/93136/how-to-plot-geo-data-using-matplotlib-python
import matplotlib.pyplot as plt 
from descartes import PolygonPatch
BLUE = '#6699cc'
fig = plt.figure() 
ax = fig.gca() 
for i in range(len(zips_data['features'])):
    if zips_data['features'][i]['properties']['ZCTA5CE10'] in atx_zip_codes:
        ax.add_patch(PolygonPatch(zips_data['features'][i]['geometry'], fc=BLUE, ec=BLUE, alpha=0.5, zorder=2 ))
ax.axis('scaled')
plt.show()
```
