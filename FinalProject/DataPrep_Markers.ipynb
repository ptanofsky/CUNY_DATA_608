```python
df_shoot = pd.read_csv('Officer_Involved_Shootings_2008-17_Incidents.csv') 
df_shoot
```

```python
df_shoot = df_shoot[['Call Type Categories','Location 1']]
df_shoot
```

```python
new = df_shoot["Location 1"].str.split("\(", n = 1, expand = True) 
new
```

```python
del new[0]
new
```

```python
new1 = new[1].str.split(" ", n = 1, expand = True) 
new1
```

```python
df_shoot["Lat"] = new1[0]
df_shoot["Long"] = new1[1]

df_shoot
```

```python
del df_shoot['Location 1']
df_shoot.head()
```

```python
df_elec = pd.read_csv('Plug-In_EVerywhere_Charging_Station_Network.csv') 
df_elec
```

```python
df_elec = df_elec[['Station Name','New Georeferenced Column']]

df_elec.head()
```

```python
df_elec['Category'] = 'ElecCarCharging'
df_elec.head()
```
